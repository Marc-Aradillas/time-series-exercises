```python
# assigning my_changable_url current url.
my_changable_url = 'https://swapi.dev/api/people/'
```

```python
def get_data():
    '''
    this function gets data from api
    '''
    
    # Initialize an empty list to store the data
    data = []
    
    # Loop through every page in the API
    for i in range(1, 10):
        # Make the API request
        response = requests.get(f'{my_changable_url}?page={i}')
        
        # Check if the response is valid
        if response.status_code == 200:
            # Get the data from the response
            data.extend(response.json()['results'])
        else:
            # If the response is not valid, break the loop
            break
    
    # Convert the list of dictionaries to a dataframe
    df = pd.DataFrame(data)

    return df
```

```python
# Print the dataframe
people = get_data()
```
